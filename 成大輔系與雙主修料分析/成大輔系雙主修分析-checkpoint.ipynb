## 0.imports module:
-------------------------------------------------


```python
import json
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib
import pylab as pl
from matplotlib.font_manager import FontProperties
from pylab import mpl
%matplotlib inline
```

## 1.取得資料和資料整理:
----------------------------------------------
> * 為了取得 ** 成功大學註冊組輔系與雙主修資料 ** , 我用pandas讀取excel檔, 取得的資料格式為 ** dataframe ** 格式
> * 因為讀取檔案的columns和資料並沒有對上, 所以我從原始資料的第二行開始讀檔, 並自己定義columns的名稱

```python
data = pd.read_excel("http://reg.acad.ncku.edu.tw/var/file/41/1041/img/2713/106dm.xlsx",header = None)
df1 = data.loc[2:]
df1.columns = ['系所代號','英文學院代號','學院','系所', '年級','輔系/雙主修','輔系/雙主修學系','人數','unnamed0','unnamed1' ]
```

> * 因為取得的資料最後有兩行全為 NAN 的資料, 所以用pandas.drop()來將它刪除
> * "系所"資料裡每個系所名稱後面還有很多空格, 因此用rstrip()將空格省略 

```python
df1 = df1.drop(['unnamed0','unnamed1'], axis=1)
df1['系所'] = df1['系所'].str.rstrip()
df1['輔系/雙主修學系'] = df1['輔系/雙主修學系'].str.rstrip()
df1['輔系/雙主修'] = df1['輔系/雙主修'].str.rstrip()
df1.head()
```

## 2(a).繪製各系輔系與雙主修人數直方圖:
----------------------------------------------
> * 利用pandas.groupby() 針對"系所"分組
> * 因為在 X軸寫入中文字可能重疊, 影響排版, 所以用英文字當索引值

```python
major = df1.groupby(['系所'])

x = list(major.groups.keys())
y = list(major.sum()['人數'])
z = []
A = 65
for  i in range(len(x)):
    z.append(chr(A))
    if A == 90:
        A=96
    A += 1
dictionary = dict(zip(z, x))
```

> * 秀出長條圖

```python
font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=14)
plt.figure(figsize = (12,6))
plt.title(u'各科系輔系與雙主修人數',fontproperties=font,fontsize=18)
plt.xlabel(u'科系',fontproperties=font,fontsize=18)
plt.ylabel(u'人數',fontproperties=font,fontsize=18)

plt.bar(z,y,width = 0.6)
for a,b in zip(z, y):
    plt.text(a, b, str(b),horizontalalignment='center',fontsize=18)

plt.show()
print (json.dumps(dictionary, indent=4,ensure_ascii=False))
```

## 2(b).結果討論:
--------------------------------------
> 由上圖可以發現**人數最多前5名**有輔系或雙主修的系是:
> * 外國語文學系學士班
> * 政治學系學士班
> * 歷史學系學士班
> * 中國文學系學士班
> * 企業管理學系學士班  
>其中前4名分別都是文學院的科系, 我推測的原因可能是:  
>1.空堂比較多 
>2.課業壓力或硬度較低 
>3.系上所學在社會上的應用可能比較侷限  
> 由上圖可以發現**人數最少前5名**有輔系或雙主修的系是:
> * 光電科學與工程學系學士班
> * 工程科學系學士班
> * 建築學系學士班
> * 物理治療學系學士班
> * 護理學系學士班  
其中分別是工學院與醫學院的科系, 我推測的原因可能是:  
>1.較少課餘時間 
>2.課業壓力或硬度較高

## 3(a).繪製各輔系與各雙主修人數直方圖:
----------------------------------------------
> * 先對 **輔系/雙主修學系** 以及 **輔系/雙主修** 進行分類
> * 雖然部分欄位因為相加總合產生亂碼, 但因為只對系所和人數作圖, 所以不會受到影響

```python
figure = df1.groupby(['輔系/雙主修學系', '輔系/雙主修']).sum()
figure.head(4)
```

> * 有部分的科系沒有人輔系或雙主修, 故需要新增欄位並將它設為0

```python
for i in x:
    if not (figure.index.isin([(i, '輔系')]).any()):
        figure.loc[(i,'輔系'),['系所代號','英文學院代號','學院','系所', '年級','人數']] = (0,0,0,0,0,0)
for i in x:
    if not (figure.index.isin([(i, '雙主修')]).any()):
        figure.loc[(i,'雙主修'),['系所代號','英文學院代號','學院','系所', '年級','人數']] = (0,0,0,0,0,0)
y1 = []
y2 = []
for i in x:
    y1.append(int(figure.loc[(i,'雙主修'),'人數']))
for i in x:
    y2.append(int(figure.loc[(i,'輔系'),'人數']))
```

> * 畫出統計圖型
> * 藍色線為雙主修人數統計, 紅色線為輔系人數統計

```python
plt.figure(figsize = (20,10))
X = np.arange(len(x))
plt.bar(X , y1, color = 'b', width = 0.35,label = 'double major',align='center')
plt.bar(X+0.2 , y2, color = 'r', width = 0.35,label = 'minor',align='edge')
plt.xticks(X,z,fontproperties=font,fontsize=16,rotation = 10)
plt.title(u'各系被輔系與雙主修人數',fontproperties=font,fontsize=18)
plt.xlabel(u'科系',fontproperties=font,fontsize=18)
plt.ylabel(u'人數',fontproperties=font,fontsize=18)
plt.legend()
for a,b in zip(X, y1):
    plt.text(a, b, str(b),horizontalalignment='center',fontsize=18)
for a,b in zip(X+0.4, y2):
    plt.text(a, b, str(b),horizontalalignment='center',fontsize=18)
plt.show()
```

## 3(b).結果討論:
----------------------------------------------
> 本系可能因為錄取成績或其他因素, 不一定是自己有興趣的科系  
> 但是選擇輔系或雙主修的科系, 必定是自己喜歡的科系  
> 根據上表, 在**輔系**中最多人選擇的是"**心理學系**", 可見心理學是許多大學生心中渴望探索的科系  
> 在**雙主修**中最多人選擇的是"**歷史學系學士班**"和"**資訊工程學系學士班**",我認為可能跟興趣和未來發展有很大的關係

## 4(a).繪製歷年輔系與雙主修人數折線圖:
----------------------------------------------
> * 將整理檔案資料的動作寫入函數, 方便修改, 也避免程式碼冗長
> * 將歷年成功大學輔系與雙主修資料爬下

```python
def set_df(data):
    df = data.loc[2:]
    df.columns = ['系所代號','英文學院代號','學院','系所', '年級','輔系/雙主修','輔系/雙主修學系','人數']
#     df['系所'] = df['系所'].str.rstrip()
#     df['輔系/雙主修學系'] = df['輔系/雙主修學系'].str.rstrip()
#     df['輔系/雙主修'] = df['輔系/雙主修'].str.rstrip()
    return df
df2 = set_df(pd.read_excel("http://reg.acad.ncku.edu.tw/var/file/41/1041/img/2713/105dm.xls",header = None))
df3 = set_df(pd.read_excel("http://reg.acad.ncku.edu.tw/var/file/41/1041/img/2713/104dm.xls",header = None))
df4 = set_df(pd.read_excel("http://reg.acad.ncku.edu.tw/var/file/41/1041/img/2713/103dm.xls",header = None))
df5 = set_df(pd.read_excel("http://reg.acad.ncku.edu.tw/var/file/41/1041/img/2713/102dm.xls",header = None))
```

> * 雖然為相同性質的歷年資料, 但是每份檔案的格式卻不相同, 所以我用不同的方式取得人數統計

```python
px=[]
px.append(int(df1.loc[df1['學院']=='總計','人數']))
px.append(int(df2.loc[df2['學院']=='加總','人數']))
px.append(int(df3.loc[df3['系所代號']=='總計','人數']))
px.append(int(df4['人數'].sum()))
px.append(int(df5['人數'].sum()))
py = range(106, 101, -1)
plt.title(u'歷年輔系與雙主修人數變化',fontproperties=font,fontsize=18)
plt.xlabel(u'學年度',fontproperties=font,fontsize=18)
plt.ylabel(u'人數',fontproperties=font,fontsize=18)

plt.plot(py,px, '-o')
plt.show()
```

## 4(b).結果討論:
----------------------------------------------
> 根據成功大學註冊組統計, 成功大學學士班總和約有11000人, 將輔系與雙主修以500人計算,    
> 也就是說有約"**5%**"的學生輔系及雙主修, 每20位學生就有1位,   
> 並且在近幾年持續上升, 維持在500人上下
